# 3. Migración a la base de datos No relacional

Para migrar la base de datos relacional (Postgres) a una base de datos no relacional (MongoDB) es importante tener en cuenta que MongoDB maneja de forma diferente las relaciones, por lo que se replantearon las relaciones existentes en la base de datos relacional para adaptarlas a MongoDB, ya que MongoDB maneja las relaciones de forma denormalizada, mediante la incorporación de la información en un solo documento. De este analisis se optó por crear 4 colecciones (jugador, equipo, estadio, universidad) dentro de la base de datos en MongoDB.

## Esquema  de datos

La razón por la cual se crean colecciones separadas para la tabla jugadores, equipos, estadios y universidades es debido a que cada uno de estos representa una entidad o un objeto de negocio diferente, y almacenarlos en colecciones separadas permite una mayor flexibilidad en la estructura de los datos, facilitando la consulta y actualización de la información.
Además, al tener estas entidades en colecciones separadas, se pueden crear índices específicos para cada una de ellas y agregar las relaciones adjuntas dentro de la colección, lo que mejoraría el rendimiento de las consultas.

````python
# formato de fecha YYYY-MM-DD

from datetime import datetime

fecha_inicio='1950-10-01' # parámetros variables
fecha_final='1970-07-01'

def convert_date(date:str):
    return datetime.strptime(date, '%Y-%m-%d')

consulta=db['estadio'].find({"$and": [
							{'partidos.primerPartido': {'$gte': convert_date(fecha_inicio)}},
							{'partidos.ultimoPartido': {'$lte': convert_date(fecha_final)}},
						]}).sort("partidos.publico", -1).limit(1)[0]
print(f"El estadio que albergó mas público para este rango de fechas fue el estadio {consulta['nombre']} "
		f"con un total de {consulta['partidos'][0]['publico']} personas")
````

### 2. Descripción
Seleccionar el top 5 de los jugadores con mayor número de jonrones en sus estadisticas de bateo

````python
from bson.son import SON
pipeline = [
    {'$unwind': '$estadisticas.bateo'},
    {'$group': {'_id': '$_id', 'count': {'$sum': '$estadisticas.bateo.HR'}}},
    {'$sort': SON([('count', -1)])},
    {'$limit': 5},
]

consulta=db['jugador'].aggregate(pipeline)
for i in consulta:
    sub_consulta = db['jugador'].find_one({"_id": i['_id']},
                                    {'jugadorID': True, 'nombreDeportivo': True})
    print(f"El jugador conocido como {sub_consulta['nombreDeportivo']} tiene un registro de "
            f"{i['count']} jonrones")
````

# 6. Actualizaciones a base de datos no relacional

### 6.1 Actualizar el valor de un campo de una tabla de la bbdd
En este ejemplo se actualizará el nombre deportivo de un jugador

````python
jugadorID="abbotku01"
nombre_deportivo_nuevo="Kiliam Boutar"
db['jugador'].update_one({"jugadorID": jugadorID},
                            {"$set": {"nombreDeportivo": nombre_deportivo_nuevo}})
print('La información del jugador fue actualizada exitosamente')

# Confirmación de la actualización del registros
confirmar = db['jugador'].find_one({"jugadorID": jugadorID},
                            {"nombreDeportivo": 1})
print(confirmar['nombreDeportivo'])
````

### 6.2 Cambiar el tipo de datos de un campo de una tabla de la bbdd.
En este ejemplo se cambiará el dato **altura** y **peso** del jugador a un campo de tipo String

````python
from pprint import pprint

db['jugador'].update_many(
    {"altura": {"$type": "int"}, "peso": {"$type": "int"}},
    [{"$set": {"altura": {"$toString": "$altura"}}},
        {"$set": {"peso": {"$toString": "$peso"}}}
    ]
)
print('Se actualizó correctamente las claves altura y peso de la colección de jugador')

# Confirmación de la actualización
confirmar = db['jugador'].find({},{"altura": 1, 'peso': 1})[0]
pprint(confirmar)
````
